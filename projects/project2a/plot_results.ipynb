```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
baseline_steps = np.array([37000, 74000, 111000, 148000])
anli_r1_baseline = np.array([0.332, 0.344, 0.364, 0.336])
anli_r2_baseline = np.array([0.344, 0.304, 0.288, 0.344])
anli_r3_baseline = np.array([0.456, 0.268, 0.272, 0.448])
hellaswag_baseline = np.array([0.288, 0.292, 0.304, 0.284])
baselines = [anli_r1_baseline, anli_r2_baseline, anli_r3_baseline, hellaswag_baseline]

steps_74000 = np.array([74000, 74000+6730, 74000+13460])
anli_r1_74000 = np.array([0.344, 0.336, 0.336])
anli_r2_74000 = np.array([0.304, 0.344, 0.344])
anli_r3_74000 = np.array([0.268, 0.388, 0.408])
hellaswag_74000 = np.array([0.292, 0.276, 0.296])
checkpoint1 = [anli_r1_74000, anli_r2_74000, anli_r3_74000, hellaswag_74000]

steps_111000 = np.array([111000, 111000+6730, 111000+13460])
anli_r1_111000 = np.array([0.364, 0.324, 0.316])
anli_r2_111000 = np.array([0.288, 0.328, 0.336])
anli_r3_111000 = np.array([0.272, 0.268, 0.216])
hellaswag_111000 = np.array([0.304, 0.292, 0.276])
checkpoint2 = [anli_r1_111000, anli_r2_111000, anli_r3_111000, hellaswag_111000]

steps_148000 = np.array([148000, 148000+6730, 148000+13460])
anli_r1_148000 = np.array([0.336, 0.332, 0.328])
anli_r2_148000 = np.array([0.344, 0.344, 0.36])
anli_r3_148000 = np.array([0.448, 0.448, 0.376])
hella_swag_148000 = np.array([0.284, 0.292, 0.284])
checkpoint3 = [anli_r1_148000, anli_r2_148000, anli_r3_148000, hella_swag_148000]
```

```python
titles = ['anli r1', 'anli r2', 'anli r3', 'hellaswag']

fig, axs = plt.subplots(2, 2)

for i in range(2) :
    for j in range(2) :
        axs[i,j].plot(baseline_steps, baselines[2*i+j], color='blue')
        axs[i,j].plot(steps_74000, checkpoint1[2*i+j], color='orange')
        axs[i,j].plot(steps_111000, checkpoint2[2*i+j], color='orange')
        axs[i,j].plot(steps_148000, checkpoint3[2*i+j], color='orange')
        axs[i,j].set_title(titles[2*i+j])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
plt.savefig('results.png')
```

```python
fig, axs = plt.subplots(1, 1)

axs.plot(baseline_steps, baselines[1], color='blue')
axs.plot(steps_74000, checkpoint1[1], color='orange')
axs.plot(steps_111000, checkpoint2[1], color='orange')
axs.plot(steps_148000, checkpoint3[1], color='orange')
axs.set_ylabel('score')
axs.set_xlabel('training iteration')
axs.set_title(titles[1])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
plt.savefig('results.png')
```
